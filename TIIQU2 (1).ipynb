```python
from pyvis.network import Network
import networkx as nx

# Create a deeply nested tree
G = nx.Graph()
G.add_edges_from([(1, 2), (1, 3), (2, 4), (2, 5), (3, 6), (3, 7), (4, 8), (4, 9)])

# Create a pyvis network
nt = Network(directed=True)

# Add nodes and edges to the network
for node in G.nodes:
    nt.add_node(node, label=f"Node {node}")
for edge in G.edges:
    nt.add_edge(edge[0], edge[1])

# Enable dynamic pop-ups when hovering over nodes
nt.show_buttons(filter_=['nodes'])
nt.set_edge_smooth('dynamic')

# Save the graph as an HTML file
nt.write_html('tree.html')
```

## XX-DYNAMIC

```python
# Create an empty graph
nt = Network(height='100%', width='100%', notebook=True, directed=True, cdn_resources='in_line')

# Iterate over each row in the DataFrame
for _, row in df.iterrows():
    topic = row['topic_name']
    words = row['topic_words'].split(', ')
    nt.add_node(topic, label=topic)
    for word in words:
        nt.add_node(word, label=word)
        nt.add_edge(topic, word)

# Generate the tree graph
nt.set_options("""
{
  "nodes": {
    "shape": "box",
    "font": {
      "size": 16,
      "color": "#000000"
    },
    "color": {
      "border": "#2B7CE9",
      "background": "#97C2FC"
    }
  },
  "edges": {
    "color": "gray"
  }
}
""")

# Show the graph
nt.show('tree2.html')
```

```python
from pyvis.network import Network
import json

# Create a network
nt = Network(height='800px', width='100%', notebook=True)

# Add nodes and edges to the network
for _, row in df.iterrows():
    topic = row['topic_name']
    words = row['topic_words'].split(', ')
    nt.add_node(topic, title=topic)
    for word in words:
        nt.add_node(word, title=word)
        nt.add_edge(topic, word)

# Set the options for the network
options = {
    "edges": {
        "color": {
            "inherit": "both",
            "highlight": "#FF0000"  # Set the highlight color for the edges
        }
    },
    "interaction": {
        "hover": True
    }
}

nt.set_options(json.dumps(options))
# Save the graph as an HTML file
nt.write_html('tree10.html')

# Show the graph
nt.show('tree.html')
```

## IN RDF

```python
import rdflib
from rdflib.extras.external_graph_libs import rdflib_to_networkx_graph
from pyvis.network import Network
import json

# Load RDF graph from a Turtle file
graph = rdflib.Graph()
graph.parse('output_10records.ttl', format='turtle')

# Convert RDF graph to NetworkX graph
nx_graph = rdflib_to_networkx_graph(graph)

# Create a pyvis Network
nt = Network(height='800px', width='100%', notebook=True, heading='RDF Graph Visualization')

# Add nodes and edges to the pyvis Network
for node in nx_graph.nodes:
    nt.add_node(str(node))
for edge in nx_graph.edges:
    nt.add_edge(str(edge[0]), str(edge[1]))

# Set the options for the network
options = {
    "edges": {
        "color": {
            "inherit": "both",
            "highlight": "#FF0000"  # Set the highlight color for the edges on hover
        }
    },
    "interaction": {
        "hover": True
    }
}

nt.set_options(json.dumps(options))

# Show the pyvis Network
nt.show('your_graph.html')
```
